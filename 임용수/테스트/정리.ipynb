```python
import os
import re

# 최상위 경로 설정
root_path = r'G:\내 드라이브\빅데이터_부산\텍스트(네이버 리뷰)'

def clean_place_name(filename):
    # 1. 확장자 분리
    name, ext = os.path.splitext(filename)
    
    # 2. 제거할 패턴들 정의
    patterns = [
        r'naver_blog_',   # 접두사
        r'_stable',       # 접미사
        r'_재추출',       # 접미사
        r'_블로그',       # 접미사
        r'_리뷰',         # 접미사
        r'복사본',         # 기타
    ]
    
    for p in patterns:
        name = re.sub(p, '', name)
    
    # 3. 언더바(_)를 공백으로 바꿨다가 모든 공백 제거 (띄어쓰기 완전 제거)
    name = name.replace('_', '').replace(' ', '')
    
    # 4. 양끝 불필요한 공백 제거 후 다시 합치기
    return name.strip() + ext

print("파일명 정제 작업을 시작합니다...")

for addr, dirs, files in os.walk(root_path):
    for file_name in files:
        if file_name.endswith(('.csv', '.xlsx')):
            # 새 파일명 생성
            new_name = clean_place_name(file_name)
            
            old_path = os.path.join(addr, file_name)
            new_path = os.path.join(addr, new_name)
            
            # 파일명이 바뀔 경우에만 실행
            if old_path != new_path:
                try:
                    # 동일한 이름의 파일이 이미 있는지 확인 (충돌 방지)
                    if not os.path.exists(new_path):
                        os.rename(old_path, new_path)
                        print(f"[변경] {file_name} -> {new_name}")
                    else:
                        print(f"[중복 제외] {new_name} 이 이미 존재하여 {file_name}은 건너뜁니다.")
                except Exception as e:
                    print(f"[오류] {file_name} 변경 실패: {e}")

print("\n모든 파일명 정리가 완료되었습니다.")
```
