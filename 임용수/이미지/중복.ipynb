```python
import os
import hashlib
from pathlib import Path

def delete_duplicate_images(folder_path):
    # Path ê°ì²´ë¡œ ë³€í™˜ (ê²½ë¡œ ìë™ êµì •)
    target_dir = Path(folder_path)
    
    if not target_dir.exists():
        print(f"âŒ ê²½ë¡œë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {target_dir}")
        # ì§„ë‹¨ì„ ìœ„í•´ ìƒìœ„ í´ë”ê°€ ì¡´ì¬í•˜ëŠ”ì§€ í™•ì¸
        if target_dir.parent.exists():
            print(f"ğŸ’¡ {target_dir.parent} í´ë”ëŠ” ì¡´ì¬í•©ë‹ˆë‹¤.")
            print(f"ğŸ“‚ ì‹¤ì œ ë‚´ë¶€ í´ë” ëª©ë¡: {os.listdir(target_dir.parent)}")
        return []

    duplicates = []
    hash_keys = dict()

    print(f"ğŸ” '{target_dir}' íƒìƒ‰ ì¤‘...")

    # í´ë” ë‚´ì˜ íŒŒì¼ë“¤ ìˆœíšŒ
    for file_path in target_dir.iterdir():
        if file_path.is_file():
            try:
                with open(file_path, 'rb') as f:
                    file_hash = hashlib.md5(f.read()).hexdigest()

                if file_hash not in hash_keys:
                    hash_keys[file_hash] = file_path.name
                else:
                    duplicates.append(file_path.name)
                    file_path.unlink()  # os.removeë³´ë‹¤ ì•ˆì „í•œ ì‚­ì œ ë°©ì‹
            except Exception as e:
                print(f"âš ï¸ ì˜¤ë¥˜ ë°œìƒ ({file_path.name}): {e}")

    return duplicates

# --- ê²½ë¡œ ì„¤ì • (ê°€ì¥ ì•ˆì „í•œ ë°©ì‹) ---
# ë¬¸ìì—´ ì•ì— rì„ ë¶™ì´ê³ , í´ë”ëª…ë“¤ì„ ì‰¼í‘œë¡œ ì—°ê²°í•˜ì„¸ìš”.
base_path = Path(r"G:\ë‚´ ë“œë¼ì´ë¸Œ\ë¹…ë°ì´í„°_ë¶€ì‚°\ì´ë¯¸ì§€1ì°¨ìˆ˜ì •")
target_name = "ìê°ˆì¹˜ì‹œì¥"

# / ì—°ì‚°ìë¥¼ ì‚¬ìš©í•˜ë©´ Path ê°ì²´ê°€ ì•ˆì „í•˜ê²Œ ê²½ë¡œë¥¼ í•©ì³ì¤ë‹ˆë‹¤.
full_path = base_path / target_name

# ì‹¤í–‰
deleted_files = delete_duplicate_images(full_path)

print("-" * 30)
print(f"âœ… ì‚­ì œëœ ì¤‘ë³µ ì‚¬ì§„ ê°œìˆ˜: {len(deleted_files)}")
```

```python
# ì´ë¯¸ì§€1ì°¨ìˆ˜ì • ë‚´ë¶€ì˜ ëª¨ë“  í´ë” ë¦¬ìŠ¤íŠ¸ì—… (desktop.ini ì œì™¸)
all_folders = [f for f in os.listdir(base_path) if os.path.isdir(base_path / f)]

for name in all_folders:
    path = base_path / name
    deleted = delete_duplicate_images(path)
    print(f"[{name}] ì‚­ì œëœ íŒŒì¼: {len(deleted)}ê°œ")
```
