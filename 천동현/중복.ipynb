```python
import os
import hashlib

def delete_duplicate_images(folder_path):
    files_list = os.listdir(folder_path)
    duplicates = []
    hash_keys = dict()

    for file_name in files_list:
        file_path = os.path.join(folder_path, file_name)
        if os.path.isfile(file_path):
            # 파일 내용으로 고유 해시값 생성
            with open(file_path, 'rb') as f:
                file_hash = hashlib.md5(f.read()).hexdigest()

            # 이미 존재하는 해시면 삭제, 아니면 딕셔너리에 추가
            if file_hash not in hash_keys:
                hash_keys[file_hash] = file_name
            else:
                duplicates.append(file_name)
                os.remove(file_path)  # 중복 파일 즉시 삭제

    return duplicates

# 학습 데이터 폴더 경로 입력
target_folder = '송정해수욕장'
deleted_files = delete_duplicate_images(target_folder)

print(f"삭제된 중복 사진 개수: {len(deleted_files)}")
for name in deleted_files:
    print(f"삭제됨: {name}")
```
